**Iris flower has three species; setosa, versicolor, and virginica, which differs according to their measurements. Now assume that you have the measurements of the iris flowers according to their species, and here your task is to train a machine learning model that can learn from the measurements of the iris species and classify them. Although the Scikit-learn library provides a dataset for iris flower classification, you can also download the same dataset from here for the task of iris flower classification with Machine Learning**

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

#Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:,2:4] #Only use petal Length and petal width
y = iris.target

#print(x)
#print(y)

target_names = iris.target_names
color = ['red','green','blue']

#Predict for new point
new_point = [[5.0,0.8]] #petal Length=6.4,width=0.2

#Create and train KNN model(k=3)
#n_neighbors default value is 5
#which one is closet value,choose that andprint it.

k=3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X,y)

#Predict and find 3 nearest neighbors
distances,indices = knn.kneighbors(new_point) #indices of k-nearest
predicted_class = knn.predict(new_point)[0]

#Plot the entire dataset
plt.figure(figsize=(8,6))
for i in range(3):
    plt.scatter(X[y==i,0],X[y==i,1],
                color=color[i],label=target_names[i],alpha=0.6)

#Highlight the 3 nearest neighbors
plt.scatter(X[indices[0],0],X[indices[0],1],
            facecolors='none',edgecolors='black',s=200,linewidths=2,label='3 Nearest Neighbors')

#Plot the new input point
plt.scatter(new_point[0][0],new_point[0][1],
            color='black',s=150,marker='X',label='New Point')

plt.title(f"KNN (k={k}) Prediction:{target_names[predicted_class]}")
plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal Width(cm)")
plt.legend()
plt.grid(True)
plt.show()
```
